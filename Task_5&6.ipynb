```python
# Assuming the last column is the target variable
X_train = train_data.iloc[:, :-1]
y_train = train_data.iloc[:, -1]
```

```python
# Split the data into training and validation sets
X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

**Decision Tree:A Decision Tree is a supervised machine learning algorithm used for both classification and regression tasks. It creates a model that makes decisions based on input features. The decision-making process involves splitting the data into subsets based on the most significant attribute at each node, resulting in a tree-like structure.**

```python
from sklearn.tree import DecisionTreeClassifier
# Create a Decision Tree model
dt_model = DecisionTreeClassifier()
```

```python
# Fit the model on the training set
dt_model.fit(X_train_split, y_train_split)
```

```python
y_val_pred = dt_model.predict(X_val_split)
```

```python
# Evaluate the model on the validation set
accuracy_val = accuracy_score(y_val_split, y_val_pred)
print(f"Accuracy on validation set: {accuracy_val*100}")
```

```python
X_test = test_data[X_train.columns]  # Assuming 'X_train.columns' contains the feature names used during training
```

```python
# Make predictions on the test set
y_test_pred = dt_model.predict(X_test)
```

```python
y_test = test_data.iloc[:, -1]
# Evaluate the model on the test set
accuracy_test = accuracy_score(y_test, y_test_pred)
print(f"Accuracy on test set: {accuracy_test*100}")
```

**SVM:SVM is effective when dealing with high-dimensional spaces and situations where a clear margin between classes is desirable. It is commonly used for classification tasks, especially in scenarios with complex decision boundaries.**

```python
# Import necessary libraries
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
```

```python
# Standardize the data (important for SVM)
scaler = StandardScaler()
X_train_split_scaled = scaler.fit_transform(X_train_split)
X_val_split_scaled = scaler.transform(X_val_split)
```

```python
# Select only numeric columns for scaling
numeric_cols = X_train.select_dtypes(include=['float64', 'int64']).columns
X_train_split_numeric = X_train_split[numeric_cols]
X_val_split_numeric = X_val_split[numeric_cols]
```

```python
# Create an SVM model
svm_model = SVC()
```

```python
# Fit and transform the scaler on the training data
X_train_split_scaled = scaler.fit_transform(X_train_split_numeric)
```

```python
# Transform the validation data using the same scaler
X_val_split_scaled = scaler.transform(X_val_split_numeric)
```

```python
# Fit the model on the scaled training set
svm_model.fit(X_train_split_scaled, y_train_split)
```

```python
X_test_numeric = test_data[numeric_cols]
```

```python
# Transform the test data using the same scaler
X_test_scaled = scaler.transform(X_test_numeric)
```

```python
# Make predictions on the scaled validation set
y_val_pred = svm_model.predict(X_val_split_scaled)
```

```python
# Evaluate the model on the validation set
accuracy_val = accuracy_score(y_val_split, y_val_pred)
print(f"Accuracy on validation set: {accuracy_val*100}")
```

```python
# Make predictions on the scaled test set
y_test_pred = svm_model.predict(X_test_scaled)
```

```python
# Evaluate the model on the test set
accuracy_test = accuracy_score(y_test, y_test_pred)
print(f"Accuracy on test set: {accuracy_test*100}")
```

```python
# Convert labels to binary format
y_train_binary = (y_train > 1).astype(int)  # Converts 2 and 3 to 1, leaves 0 and 1 unchanged
y_test_binary = (y_test > 1).astype(int)
```

```python
unique_labels = np.unique(y_train_binary)
print("Unique labels in y_train:", unique_labels)
```

```python
# Convert data to DMatrix format, which is a specific data structure used by XGBoost
dtrain = xgb.DMatrix(X_train, label=y_train_binary)
dtest = xgb.DMatrix(X_test, label=y_test_binary)
```

```python
# Define parameters for XGBoost
params = {
    'max_depth': 3,
    'learning_rate': 0.1,
    'objective': 'binary:logistic',  # For binary classification
    'eval_metric': 'logloss'
}
```

```python
# Train the XGBoost model
num_rounds = 10  # You can adjust the number of boosting rounds based on your data
bst = xgb.train(params, dtrain, num_rounds)
```

```python
# Make predictions on the test set
y_pred = bst.predict(dtest)
predictions = [round(value) for value in y_pred]
```

```python
# Evaluate the accuracy of the model
accuracy = accuracy_score(y_test, predictions)
print(f"Accuracy: {accuracy * 100:.2f}%")
```
