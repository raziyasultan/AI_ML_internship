**Univariate analysis refers to the statistical analysis of a single variable at a time. It is the simplest form of statistical analysis because it deals with only one variable. The main purpose of univariate analysis is to describe the distribution of data and to understand the central tendency and dispersion of a variable.**

```python
df.columns
```

**Let us Consider City column to analyze the univariate analysis**

```python
categorical_data = df['Make']
```

```python
value_counts = categorical_data.value_counts()
```

```python
value_counts
```

```python
plt.figure(figsize=(25, 10))
sns.countplot(x=categorical_data, palette='viridis')
plt.title('Distribution of Categorical Variable')
plt.xlabel('Categories')
plt.ylabel('Frequency')
plt.show()
```

```python
percentage_distribution = value_counts / len(categorical_data) * 100
print("Percentage Distribution:")
print(percentage_distribution)
```

```python
plt.figure(figsize=(10,10))
plt.pie(percentage_distribution, labels=percentage_distribution.index, autopct='%1.1f%%', colors=sns.color_palette('viridis'))
plt.title('Percentage Distribution of Categorical Variable')
plt.show()
```
